```python

import pickle
import gzip
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import softmax
import pandas as pd    
import torch
import numpy as np

# exponential increase in the prob of a bad sample. 
```

```python
times = np.arange(200)

def probDerailed(time, good_prob):
    return 1-(( good_prob )**time)

for bad_prob in [0.01, 0.05, 0.1, 0.2]:
    #bad_prob = 0.01
    good_prob = 1 -bad_prob

    derail_probs = []
    for t in times: 
        derail_probs.append(probDerailed(t, good_prob))  
    plt.plot(times, derail_probs, label='Bad Token Cum. Prob: '+str(bad_prob))
    
plt.title('Prob of sampling at Least One Bad Token')
plt.xlabel('Time Point')
plt.ylabel('Probability')
plt.legend()
plt.gcf().savefig('ImageOutputs/Prob_of_derailing.png', dpi=250)
```
